```python
aggregated_data.loc[:,'session_duration'] = (aggregated_data['session_end'] - aggregated_data['session_start']).dt.total_seconds() / 60
```

```python
user_base_session = aggregated_data.groupby('user_id').agg(
    num_sessions=('session_id', 'nunique'),
    sum_page_clicks=('page_clicks', 'sum'),
    avg_session_duration=('session_duration', 'mean')
).reset_index()
```

```python
Not_cancelled_trips['session_start'] = pd.to_datetime(Not_cancelled_trips['session_start'])
Not_cancelled_trips['session_end'] = pd.to_datetime(Not_cancelled_trips['session_end'])
Not_cancelled_trips['departure_time'] = pd.to_datetime(Not_cancelled_trips['departure_time'])

# Filter non-cancelled trips
\

# Compute required columns
Not_cancelled_trips['num_flights'] = np.where(
    (Not_cancelled_trips['return_flight_booked']) & (Not_cancelled_trips['flight_booked']),
    2,
    np.where((Not_cancelled_trips['return_flight_booked']) | (Not_cancelled_trips['flight_booked']), 1, 0)
)
Not_cancelled_trips['hotel_expenses'] = Not_cancelled_trips['hotel_per_room_usd'] * Not_cancelled_trips['nights'] * Not_cancelled_trips['rooms']
Not_cancelled_trips['hotel_expenses'] *= np.where(Not_cancelled_trips['hotel_discount'], 1 - Not_cancelled_trips['hotel_discount_amount']/100, 1)
Not_cancelled_trips['flight_fare'] = Not_cancelled_trips['base_fare_usd'] * (1 - Not_cancelled_trips['flight_discount_amount']/100)
Not_cancelled_trips['time_after_booking'] = (Not_cancelled_trips['departure_time'] - Not_cancelled_trips['session_end']).dt.total_seconds() / 3600

# Haversine formula for distance calculation
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in kilometers
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c

Not_cancelled_trips['distance_flown'] = haversine(
    Not_cancelled_trips['home_airport_lat'],
    Not_cancelled_trips['home_airport_lon'],
    Not_cancelled_trips['destination_airport_lat'],
    Not_cancelled_trips['destination_airport_lon']
)

# Group by 'user_id' and calculate aggregates
user_base_trips = Not_cancelled_trips.groupby('user_id').agg(
    num_trips = ('trip_id','count'),
    num_flights=('num_flights', 'sum'),
    hotel_expenses=('hotel_expenses', 'sum'),
    flight_fare=('flight_fare', 'sum'),
    avg_time_after_booking=('time_after_booking', 'mean'),
    avg_km_flown=('distance_flown', 'mean'),
    avg_bags=('checked_bags', 'mean')
).reset_index()

# Display the resulting DataFrame
print(user_base_trips)
```
